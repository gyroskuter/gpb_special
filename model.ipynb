```python
import pandas as pd
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_pickle('data/modified_bank_data.pickle')
df
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
```

```python
lr_pred = lr.predict(X_test_scaled)
```

```python
print('---------------------------Metrics--------------------------')
print(f'Mape: {mean_absolute_percentage_error(y_test, lr_pred)}')
print(f'R2: {r2_score(y_test, lr_pred)}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
lgbm = LGBMRegressor(objective='huber', n_estimators=1000)
lgbm.fit(X_train, y_train)  # без скейлера
lgbm_pred = lgbm.predict(X_test)

print('LightGBM:')
print(f'MAPE: {mean_absolute_percentage_error(y_test, lgbm_pred) * 100:.2f}%')
print(f'R2: {r2_score(y_test, lgbm_pred):.4f}')
```

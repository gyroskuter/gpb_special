```python
import pandas as pd
```

```python
df = pd.read_pickle('data/bank_data.pkl')
```

```python
df_modified = df.groupby('day')['balance'].last().reset_index()
df_modified['lag_1'] = df_modified['balance'].shift(1)
df_modified['lag_7'] = df_modified['balance'].shift(7)
df_modified['lag_30'] = df_modified['balance'].shift(30)
```

```python
df_modified['rolling_mean_7'] = df_modified['balance'].rolling(7).mean()
df_modified['rolling_std_7'] = df_modified['balance'].rolling(7).std()
df_modified['rolling_mean_30'] = df_modified['balance'].rolling(30).mean()
df_modified['rolling_std_30'] = df_modified['balance'].rolling(30).std()
```

```python
df_modified['is_salary_day'] = df_modified['day'].apply(lambda x: 1 if x % 30 in [0, 15] else 0)
df_modified['is_tax_day'] = df_modified['day'].apply(lambda x: 1 if x % 30 == 28 else 0)
```

```python
df_modified['target'] = df_modified['balance'].shift(-1)  # баланс следующего дня
df_modified = df_modified.dropna()
df
```

```python
df_modified['total_in'] = df[df['direction'] == 'in'].groupby('day')['amount'].sum()
df_modified['total_out'] = df[df['direction'] == 'out'].groupby('day')['amount'].sum()
```

```python
df_modified['n_transactions'] = df['day'].value_counts()
```

```python

df_modified.to_pickle('data/modified_bank_data.pickle')
```

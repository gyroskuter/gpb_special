```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
df = pd.DataFrame(columns=['day', 'type', 'direction', 'amount', 'balance'])
transaction_types = ['CLIENT_IN', 'DEPOSIT_IN', 'CLIENT_OUT', 'LOAN_ISSUE']
direction_map = {
    'CLIENT_IN': 'in', 'DEPOSIT_IN': 'in',
    'CLIENT_OUT': 'out', 'LOAN_ISSUE': 'out',
    'SALARY': 'out', 'TAX': 'out'
}

balance = 50_000_000  # начальный остаток корсчёта
rows = []

np.random.seed(666)
for day in range(365):
    todays_transactions = [np.random.choice(transaction_types, p=[0.4, 0.1, 0.4, 0.1]) for _ in range(np.random.randint(15, 25))]
    if day % 30 == 15: todays_transactions.append('SALARY')
    elif day % 30 == 0: todays_transactions.append('SALARY')
    elif day % 30 == 28: todays_transactions.append('TAX')

    for transaction in todays_transactions:
        if transaction == 'CLIENT_IN':
            amount = np.random.uniform(100_000, 5_000_000)
        elif transaction == 'DEPOSIT_IN':
            amount = np.random.uniform(500_000, 10_000_000)
        elif transaction == 'CLIENT_OUT':
            amount = np.random.uniform(100_000, 5_000_000)
        elif transaction == 'LOAN_ISSUE':
            amount = np.random.uniform(1_000_000, 20_000_000)
        elif transaction == 'SALARY':
            amount = np.random.uniform(5_000_000, 15_000_000)
        elif transaction == 'TAX':
            amount = np.random.uniform(3_000_000, 10_000_000)

        direction = direction_map[transaction]
        balance = balance + amount if direction == 'in' else balance - amount
        balance = max(10_000_000, balance)

        rows.append({
            'day': day,
            'type': transaction,
            'direction': direction,
            'amount': round(amount, 2),
            'balance': round(balance, 2)
        })

df = pd.DataFrame(rows)
print(df.shape)
print(df.head(20))
```

```python
df.to_pickle('data/bank_data.pkl')
conn = sqlite3.connect('data/bank_data.db')
df.to_sql('transactions', conn, if_exists='replace', index=False)
conn.close()
```
